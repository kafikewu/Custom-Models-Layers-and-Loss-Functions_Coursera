```python
import numpy as np
import tensorflow as tf

# PREDEFINED

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
# labels. relationship with the inputs above is y=2x-1.
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
def custom_loss(y_true, y_pred):
  errorSquared = tf.reduce_mean(tf.square(y_true - y_pred))
  error = tf.sqrt(errorSquared)
  return error 
```

```python
## RANDOM MODEL

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 32, input_shape = (1, ), activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1))
model.compile(optimizer = tf.keras.optimizers.RMSprop(), loss = custom_loss)
```

```python
model.fit(xs, ys, epochs = 1000)
```

```python
model.predict([11.0])
```
